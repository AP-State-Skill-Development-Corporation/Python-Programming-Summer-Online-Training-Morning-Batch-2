```python
import numpy as np
x = np.linspace(1,10,100)
y = np.log(x)

plt.figure(1)

plt.subplot(2,2,1)
plt.title("y=log(x)")
plt.plot(x,y)

plt.subplot(2,2,2)
plt.title("y=log(x)**2",color="pink")
plt.plot(x,y**2)
plt.show()

plt.subplot(2,2,1)
plt.title("y=log(x)**3",color="red")
plt.plot(x,y)

plt.subplot(2,2,2)
plt.title("y=log(x)**4",color="yellow")
plt.plot(x,y**2)
plt.show()
```
